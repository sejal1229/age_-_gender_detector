```python
def Convolution(input_tensor, filters):
    x = Conv2D(filters = filters, kernel_size = (3, 3), padding = "same", strides = (1, 1), kernel_regularizer = l2(0.001))(input_tensor)
    x = Dropout(0.1)(x)
    x = Activation('relu')(x)

    return x
```

```python
def model(input_shape):
    inputs = Input((input_shape))

    ## Convolutional Layers & MaxPool Layers
    conv_1 = Convolution(inputs, 32)
    maxp_1 = MaxPooling2D(pool_size = (2, 2))(conv_1)
    conv_2 = Convolution(maxp_1, 64)
    maxp_2 = MaxPooling2D(pool_size = (2, 2))(conv_2)
    conv_3 = Convolution(maxp_2, 128)
    maxp_3 = MaxPooling2D(pool_size = (2, 2))(conv_3)
    conv_4 = Convolution(maxp_3, 256)
    maxp_4 = MaxPooling2D(pool_size = (2, 2))(conv_4)

    ## 3D matrix into 1D vector
    flatten = Flatten()(maxp_4)

    dense_1 = Dense(64, activation = 'relu')(flatten)
    dense_2 = Dense(64, activation = 'relu')(flatten)

    drop_1 = Dropout(0.2)(dense_1)
    drop_2 = Dropout(0.2)(dense_2)

    output_1 = Dense(1, activation = 'sigmoid', name = 'sex-out')(drop_1)
    output_2 = Dense(1, activation = 'relu', name = 'age-out')(drop_2)

    model = Model(inputs = [inputs], outputs = [output_1, output_2])
    model.compile(loss = ["binary_crossentropy", "mae"], optimizer = "Adam", metrics = ["accuracy"])

    return model
```

```python
import keras
print(keras.__version__)
from keras.utils import to_categorical



# Assuming you have a directory of images to load
datagen = ImageDataGenerator(rescale=1./255)  # Create ImageDataGenerator with rescale

# Load a single image for demonstration (optional)
# image = load_img('path/to/image.jpg')  # Adjust path

train_generator = datagen.flow_from_directory(
    'utkface_aligned_cropped',  # Directory with images
    target_size=(48, 48),
    color_mode='grayscale',
    class_mode='categorical',
    batch_size=64,
    shuffle=True
)

# Create X_train and Y_train_2 from the generator, using a more efficient approach
X_train, Y_train_2 = [], []
for _ in range(len(train_generator)):
    X_batch, Y_batch = next(train_generator)
    X_train.extend(X_batch)
    Y_train_2.extend(Y_batch)

# Convert X_train and Y_train_2 to NumPy arrays
X_train = np.array(X_train)
Y_train_2 = np.array(Y_train_2)

# Ensure correct data types and shapes
X_train = X_train.astype('float32')
Y_train_2 = to_categorical(Y_train_2)

# Reshape for CNN input
X_train = X_train.reshape(X_train.shape[0], 48, 48, 1)

print("X_train shape:", X_train.shape)
print("Y_train_2 shape:", Y_train_2.shape)

# Assuming your model has two outputs
metrics = ['accuracy', 'accuracy']
```

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Input, Dense

input_shape = (X_train.shape[1], X_train.shape[2], X_train.shape[3])  # Get input shape from your data

# Create the model
model = Sequential()
model.add(Input(shape=input_shape))  # Define input layer with shape
model.add(Dense(units=64, activation='relu'))  # Add layers without input_shape
model.add(Dense(units=2, activation='softmax'))  # Output layer for 2 classes

# Compile the model
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy', 'accuracy']) 
```

```python
# Then call fit with validation data and epochs
history = model.fit(X_train, Y_train_2, batch_size=128, validation_data=(X_test, Y_test_2), epochs=30,  callbacks=callback_list)
```
